## 元字符

`.` 除了换行符以外的任意字符

`[^]` 否定字符集

`^` 开始行开始匹配

`$` 从末端开始匹配


```python
import re 

pattern = re.compile(r"^(T|t)he\i")
text_1 = "The car is parked in the garage \n "
text_2 = "the car is parked in the garage \n the test"
print(re.findall(pattern,text_1))
print(re.findall(pattern,text_2))
```

## 简写字符集

大写代表`^` 

|简写| 描述|
|----|-----------|
| . | 换行符以外的字符|
|\w| 字母数字|
|\d| 数字|
|\s| 空格字符，等同于 `[\t\n\f\r\p{Z}]`|


### 零宽度断言
判断匹配的格式是否在另一个确定的格式之前，匹配结果不包含该确定的格式

|符号| 描述|
|----|---------|
|？=| 正先行-存在|
|？!| 负先行-排除|
|？<=| 正后发-存在|
|?<!| 负后发-排除|

正先行表示前一部分表达式后必须跟着`?=`定义的表达式
